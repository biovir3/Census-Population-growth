```python
county_shape = './cb_2017_us_county_5m.shp'
data = gpd.read_file(county_shape)
#plt.axis(-180,-50,0,90)

data
```

```python
census_data = pd.read_csv('./DEC_00_SF1_GCTPH1.US05PR_with_ann.csv', encoding='cp437', skiprows=1)
```

```python
merged = data.merge(census_data, left_on='AFFGEOID', right_on='Target Geo Id', how = 'inner')
#merged.info()
```

```python
fig, ax = plt.subplots(figsize = (30,30))

ax.set_title('This Visualization does not tell anything, it uses the cmap="flag" color mapping, but it looks interesting', fontsize = 'xx-large')
ax.set_xlim([-126,-66])
ax.set_ylim([23,50])
merged.plot(column = 'Population', figsize=(30,30), ax=ax, cmap='flag', vmin = 0, vmax = 750000);

```
